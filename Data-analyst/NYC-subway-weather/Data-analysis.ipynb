### Pearson correlation from scratch

```python
def correlation(x,y):
    z_x = (x - x.mean()) / x.std(ddof=1)
    z_y = (y - y.mean()) / y.std(ddof=1)
    
    return (z_x * z_y).mean()
```

Correlatation between number of people who ride the subway each hour and the mean precipitation for the day

```python
correlation(data["ENTRIESn_hourly"],data["meanprecipi"])
```

Correlatation between number of people who ride the subway each hour and a cumulative total of how many people have ridden the subway

```python
correlation(data["ENTRIESn_hourly"],data["ENTRIESn"])
```
