```python
import numpy as np
import pandas as pd
```

```python
def max_value(data):
    '''
    Fill in this function to return the name of the country
    with the highest employment in the given employment
    data, and the employment in that country.
    '''
    i = data["mean"].argmax()

    return (data["Country"][i], data["mean"][i])
```

```python
def min_value(data):
    '''
    Fill in this function to return the name of the country
    with the lowest employment in the given employment
    data, and the employment in that country.
    '''
    i = data["mean"].argmin()

    return (data["Country"][i], data["mean"][i])
```

Working with employment data

```python
#loading employment data
employment_data = pd.read_csv("employment_above_15.csv", sep=",")

employment_data
```

```python
#adding min, mean and max columns to the employment data table
employment_data["min"] = employment_data.min(axis = 1)
employment_data["mean"] = employment_data.mean(axis = 1)
employment_data["max"] = employment_data.max(axis = 1)
employment_data
```

```python
max_value(employment_data)
```

```python
min_value(employment_data)
```

```python
employment_data.sort_values(by=['mean'], ascending=False).head(10)
```

Comparing number of countries that are both above/below the means versus differ in means

```python
def similarity(data1,data2):
    above = (data1["mean"] > data1["mean"].mean()) & (data2["mean"] > data2["mean"].mean())
    below = (data1["mean"] < data1["mean"].mean()) & (data2["mean"] < data2["mean"].mean())
    
    similar = above | below
    total_similar = similar.sum()
    
    total_diff = len(data2.dropna()) - total_similar
    
    return(total_similar,total_diff)
```

```python
similarity(employment_data,gdp_data)
```

```python
gdp_data = gdp_data.set_index("Country")
gdp_data
```

Maximum value by index

```python
def max_gdp(data):
    max_country = data["mean"].argmax()
    max_value = data["mean"].loc[max_country]
    
    return max_country,max_value
```

```python
max_gdp(gdp_data)
```
