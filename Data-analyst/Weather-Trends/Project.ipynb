```python
import pandas as pd
import numpy as np
import seaborn as sns
```

Retrieve weather data

```python
doc1 = "city_list.csv"
doc2 = "city_data.csv"
doc3 = "global_data.csv"
```

First, we want to find a city close to **Toulouse, France**. We load `city_data` and select all French cities available for study

```python
city_list = pd.read_csv(doc1, sep=",")
#city_list
french_city_list = city_list[city_list["country"] == "France"]
french_city_list
```

The only French city availble is **Paris**. We will further contrast global weather trends with weather trends in the capital of France. 

We will load `global_data` as well as the data for Paris from `city_data`

```python
global_data = pd.read_csv(doc3, sep=",")
global_data
```

```python
city_data = pd.read_csv(doc2, sep=",")
#city_data
paris_data = city_data[city_data["city"] == "Paris"]
paris_data
```

We can see some `Nan` values in `Paris_data` table, we will drop them from both tables.

We will rename `avg_temp` in both tables with explicit column names

```python
global_data = global_data.dropna()
global_data.rename(columns={'avg_temp': 'global_temp'}, inplace=True)
global_data
```

We'll drop `city` and `country` columns from `paris_data` as well

```python
paris_data = paris_data.dropna()
paris_data.rename(columns={'avg_temp': 'paris_temp'}, inplace=True)
paris_data = paris_data.drop(["city","country"],axis=1)
paris_data
```

We will now merge do data sets into one `global_data` table based on year

```python
all_data = paris_data.merge(global_data, how="outer", on="year")
all_data = all_data.dropna()
all_data
```

We now can plot both Paris and global temparatures

```python
all_data = all_data.melt('year', var_name='Data',  value_name='temp')
g = sns.lineplot(x="year", y="temp", hue='Data', data=all_data)
```

However, even if we see that overall Paris temperatures are higher and global temperatures are going up, not all the trends can be clearly seen, due to fluctuations. We could get a better understanding of data if we smooth it with moving average, i.e. for the period of 10 years.

```python
all_data = paris_data.merge(global_data, how="outer", on="year")
all_data = all_data.dropna()
all_data
```

```python
all_data["paris_avg"] = all_data["paris_temp"].rolling(10).mean()
all_data["global_avg"] = all_data["global_temp"].rolling(10).mean()
all_data = all_data.drop(["paris_temp","global_temp"],axis=1)
all_data
```

```python
all_data = all_data.melt('year', var_name='Data',  value_name='avg temp')
g = sns.lineplot(x="year", y="avg temp", hue='Data', data=all_data).set_title("Weather Trends")
```
