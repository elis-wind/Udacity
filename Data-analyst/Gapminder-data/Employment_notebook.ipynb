```python
import numpy as np
import pandas as pd
```

```python
#loading employment data
employment_data = pd.read_csv("employment_above_15.csv", sep=",")

employment_data
```

```python
#adding min, mean and max columns to the employment data table
employment_data["min"] = employment_data.min(axis = 1)
employment_data["mean"] = employment_data.mean(axis = 1)
employment_data["max"] = employment_data.max(axis = 1)
employment_data
```

```python
def max_employment(data):
    '''
    Fill in this function to return the name of the country
    with the highest employment in the given employment
    data, and the employment in that country.
    '''
    i = data["mean"].argmax

    return (data["Country"][i], data["mean"][i])
```

```python
max_employment(employment_data)
```

```python
def min_employment(data):
    '''
    Fill in this function to return the name of the country
    with the lowest employment in the given employment
    data, and the employment in that country.
    '''
    i = data["mean"].argmin

    return (data["Country"][i], data["mean"][i])
```

```python
min_employment(employment_data)
```
