```python
import numpy as np
import pandas as pd
```

```python
male_data = pd.read_csv("male_completion_rate.csv")
male_data
```

```python
female_data = pd.read_csv("female_completion_rate.csv")
female_data
```

```python
male_data["mean"] = male_data.mean(axis=1)
male_data
```

```python
female_data["mean"] = female_data.mean(axis=1)
female_data
```

```python
def overall_completion_rate(female_completion, male_completion):
    '''
    Fill in this function to return a NumPy array containing the overall
    school completion rate for each country. 
    '''
    overall_data = pd.DataFrame()
    overall_data["country"] = male_completion["Country"]
    overall_data["mean"] = (male_completion["mean"] + female_completion["mean"]) / 2.
    
    return overall_data[["country","mean"]]
```

```python
overall_completion_rate(female_data, male_data)
```
